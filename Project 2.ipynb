# Project 2 Preparing Data

A dataset used on this project contrains information about 160 countries with ~40 characteristics such as debt, electricity consumption, internet users an many other characteristics. 

The data analysis will aim to answer the following questions.

### 1st dataset - Country data
 * What is the country with most consuption of electricity?
 * Top 5 countries with the highest GDP
 * What are the top 10 poorest countries?
 
Some of the challenges to ovecome during the data preparion include:
    * Ability to properly transform the data into csv format.
    * Identify missing/null information in the data set.
    
 

```python
import pandas as pd

df = pd.read_csv("https://perso.telecom-paristech.fr/eagan/class/igr204/data/factbook.csv", sep=";")
df.columns = df.columns.str.replace('-', '').str.strip().str.replace(' ', '_').str.replace('__', '_').str.replace('&', 'n').str.replace('\(kWh\)', '')
```

```python
df.head()
```

### Select wanted columns.

```python
df_n = df[["Country", "Debt_external", "GDP","Reserves_of_foreign_exchange_n_gold", "Telephones_main_lines_in_use", "Telephones_mobile_cellular", "Electricity_consumption", "Electricity_production"]]
```

```python
df_n.head()
```

### Remove entries that contain NaN values in any column

```python
df_n = df_n.dropna()
df_n = df_n.drop([0])
df_n.head()
```

### Find the country with the highest electricity consumption
richest country is the one with the minumum onternal debt and highest gold reserve and GDP

```python
df_n[df_n.Electricity_consumption == df_n.Electricity_consumption.max()]
```

### Top 5 countries with the highest GDP

```python
df_n.sort_values("GDP", ascending = False, inplace = True)
df_n.head()
```

### Country with the least number of active mobile phones

```python
df_n[df_n.Telephones_mobile_cellular == df_n.Telephones_mobile_cellular.min()]
```

### Top 10 Poorest Countries.

```python
df_n.sort_values("Debt_external", ascending = False, inplace = True)
```

```python
df_n["Debt_external"].astype(str).astype(float).nlargest(10)
```
